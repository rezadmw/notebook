### Number of sibling/spouse/children problem

From the information above, there are SibSp and Parch column. The number of the sibling/spouse/children onboard may have an effect on the survival rate. But, the greater number sibling/spouse/children doesn't mean the passanger had a greater survival chance. I will prove it by making a **correlation** analysis in the next section. 

Therefore, I'm going to add another criteria to indicate if the passanger had a relatives onboard in the next step.  

## 2. Data Cleaning & Transformation

Data Encoding

```python
labelenc = LabelEncoder()
data_enc = data.copy()
data_enc['Embarked'] = data_enc['Embarked'].fillna('No info')
data_enc['Sex'] = labelenc.fit_transform(data_enc['Sex'])
data_enc['Embarked'] = labelenc.fit_transform(data_enc['Embarked'])
```

Add 'has_relative' column

```python
data_rel = data_enc.copy()
```

```python
data_rel.insert(len(data_rel.columns), 'has_relative',np.nan)
```

```python
for i in range(len(data_rel)):
    if (data_rel['SibSp'][i+1]>0 or data_rel['Parch'][i+1]>0):
        data_rel['has_relative'][i+1] = 1
    else:
        data_rel['has_relative'][i+1] = 0
```

```python
data_rel
```

```python
corr = data_rel.corr()
fig, ax = plt.subplots(figsize=(10,10))
ax = sns.heatmap(
    corr, 
    vmin=-1, vmax=1, center=0,
    cmap=sns.diverging_palette(20, 220, n=200),
    square=True,
    
)   
ax.set_xticklabels(
    ax.get_xticklabels(),
    rotation=45,
    horizontalalignment='right'
);
```

```python
corr
```

From the number of relatives problem above, here are the correlation value that i want to prove:

Survived-SibSp = -0.0353

As you can see, the old column(SibSp and Parch) has a very close to zero value. Which means in pearson correlation, almost not related. But the **has_relatives** column has a coefficient of 0.203 which some practician say is **related but not strong**.

### Find out missing value

```python
data.isna().sum()
```

From information above, columns that has null value is **age**, **cabin**, and **embarked**. 

***
Fare
***
There are some 0 value in fare data. My assumption is that this was missing data because, every ticket must have a value. So, i'm going to try to impute fare data using multiple linear regression. 

```python
imp_fare_data = data_rel[['Pclass', 'SibSp', 'Parch','Fare']].copy()
imp_fare_data = imp_fare_data.loc[imp_fare_data['Fare']>0]
no_fare_data = data_rel.loc[data_rel['Fare']==0]
```

```python
no_fare_data = no_fare_data.reset_index(drop=False)
```

```python
list_nfd = np.array(no_fare_data['PassengerId']) #list of the PassengerId that has 0 fare
```

```python
X = np.array(imp_fare_data.iloc[:, :-1])
y = np.array(imp_fare_data.iloc[:, -1])
pred_inp = np.array(no_fare_data[['Pclass','SibSp','Parch']])
```

```python
ct1 = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [0])], remainder='passthrough')

X = np.array(ct1.fit_transform(X))
pred_inp = np.array(ct1.fit_transform(pred_inp))

print(X[0])
print(pred_inp[0])
```

```python
def mean_absolute_percentage_error(y_true, y_pred): 
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)

# Training the Multiple Linear Regression model on the Training set
regressor = LinearRegression()
regressor.fit(X_train, y_train)

# Predicting the Test set results
y_pred = regressor.predict(X_test)
np.set_printoptions(precision=2)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
mean_absolute_percentage_error(y_test,y_pred)
```

```python
fare_pred = regressor.predict(pred_inp)
len(fare_pred)
```

```python
for i in range(len(fare_pred)):
    data_rel['Fare'][list_nfd[i]] = fare_pred[i]
    data['Fare'][list_nfd[i]] = fare_pred[i]
    print(data_rel['Fare'][list_nfd[i]])
```

**Cabin**

The cabin column has a large number of missing value (>75%) which is not ideal. Even if we try to impute it, it's still hard to tell the correctness of the data. Therefore, I decided to drop the cabin column and not use it for analysis.

```python
data_rel = data_rel.drop(columns='Cabin')
```

***

**Age**

***
For the age column, the number of missing data is considerably big(**19%**). Some practician say it isn't ideal but some say it's tolerable. In this case, i will try to **impute** the missing age data first based on the data that is related to the age column. To find the data that is related to age, i will use `corr` function to find the correlation between each data.  

From the correlation analysis (using pearson correlation) we can see the correlation coefficient for each data to the other. For the age data, The biggest correlation coefficient is with Pclass and SibSp.

To impute the data, i will use a multiple linear regression. We will see if the result is good enough to fill the missing value.

```python
imp_age_data = data_rel[['Age','Sex','Fare','has_relative','Survived']].copy()
```

```python
imp_age_data = imp_age_data.reset_index(drop=True)
imp_age_data = imp_age_data.dropna() #drop missing value
```

```python
X = np.array(imp_age_data.iloc[:, 1:len(imp_age_data.columns)])
y = np.array(imp_age_data.iloc[:, 0]) 
print(X)
print(y)
```

```python
ct1 = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [0])], remainder='passthrough')

ct2 = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [3])], remainder='passthrough')

ct3 = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [5])], remainder='passthrough')

X = np.array(ct1.fit_transform(X))
X = np.array(ct2.fit_transform(X))
X = np.array(ct3.fit_transform(X))
print(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)

# Training the Multiple Linear Regression model on the Training set

regressor = LinearRegression()
regressor.fit(X_train, y_train)

# Predicting the Test set results
y_pred = regressor.predict(X_test)
np.set_printoptions(precision=2)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
mean_absolute_percentage_error(y_test,y_pred)
# plt.plot(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

The mean absolute percentage error shows that it's hard to predict the age of the passanger. Therefore, we still able to use the age data, but we have to ignore the missing value.

## 4. Conclusion

Let' try to conclude our analysis result by looking at the main goals:

- find out who had the best chance to survived.

- find out any factors that affect the survival rate. 

From the data, the passenger that had a better survival rate is :

    1. a female

    2. staying in the first or second class

    3. aged between 0-15

    4. Had relatives onboard

From that we know that age, gender, class, and relatives has a high effect on survival rate of the passanger. 

Almost all column involved to determine the survival rate of the passanger survival rate. The column that didn't relate to the survival chance are embarkation.
